```python
with open('kieu-1872.txt') as f:
    data = f.read()
```

```python
import re
def nom_only(s):
    nom_string = re.sub(r"[0-9A-ZÀÁẠẢÃÂẦẤẬẨẪĂẰẮẶẲẴÈÉẸẺẼÊỀẾỆỂỄÌÍỊỈĨÒÓỌỎÕÔỒỐỘỔỖƠỜỚỢỞỠÙÚỤỦŨƯỪỨỰỬỮỲÝỴỶỸĐa-zàáạảãâầấậẩẫăằắặẳẵèéẹẻẽêềếệểễìíịỉĩòóọỏõôồốộổỗơờớợởỡùúụủũưừứựửữỳýỵỷỹđ.,:;\-\[\]?\'\"! \t\n“”/]+", "", s)
    return list(nom_string)
```

```python
data = nom_only(data)
```

```python
from collections import Counter
```

```python
counter = Counter(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(counter.items(), columns=['label', 'Count'])
df.Count.plot(kind='hist', bins=32)

plt.xlabel('Số lần xuất hiện')
plt.ylabel('Số ký tự')
plt.yscale('log')
plt.grid()
# plt.show()
plt.savefig("word_nom_histogram.png", dpi=300)
```
